```python
# Import necessary libraries
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
```

```python
# Load your dataset
df = pd.read_csv('D:\wael gomaa21-4-2024\My Tasks\FinalProject\My project\Banking Customer Churn Prediction Dataset\Churn_Modelling.csv')
df.head()
```

```python
# Prepare the features (X) and target (y)
# Include CustomerId in the dataset for later use in Power BI
X = df[['CustomerId', 'CreditScore', 'Age', 'Tenure', 'Balance', 'NumOfProducts', 'HasCrCard', 'IsActiveMember', 'EstimatedSalary']]
y = df['Exited']  # Target variable: churn status (Exited column)

# Split the data into training and testing sets (70% training, 30% testing)
X_train, X_test, y_train, y_test = train_test_split(X.drop(columns=['CustomerId']), y, test_size=0.3, random_state=42)
```

```python
# Create and train the RandomForestClassifier model
model = RandomForestClassifier(random_state=42)
model.fit(X_train, y_train)

# Predict churn on the entire dataset (using the model)
df['PredictedChurn'] = model.predict(X.drop(columns=['CustomerId']))  # Predicted churn (0 or 1)

# Get the probability of churn
df['ProbabilityOfChurn'] = model.predict_proba(X.drop(columns=['CustomerId']))[:, 1]  # Probability of churn

# Export the result as a CSV file including CustomerId, PredictedChurn, and ProbabilityOfChurn
df[['CustomerId', 'PredictedChurn', 'ProbabilityOfChurn']].to_csv('D:\wael gomaa21-4-2024\My Tasks\FinalProject\My project\Banking Customer Churn Prediction Dataset\Churn_Prediction_Results.csv', index=False)

# Optional: Evaluate the model on the test data and print accuracy score
y_test_pred = model.predict(X_test)
accuracy = accuracy_score(y_test, y_test_pred)
print(f'Accuracy of the churn prediction model: {accuracy * 100:.2f}%')
```
